```python
from langchain_openai import AzureChatOpenAI
from langchain_community.callbacks import get_openai_callback
from langchain_openai import OpenAI
```

```python
import os
from langchain.schema import HumanMessage, SystemMessage

def image_class(image_urls):
    """Make image summary for all images by passing their public URLs."""
    chat = AzureChatOpenAI(model="firsttest", max_tokens=2000, temperature=0)

    system_message = SystemMessage(
        content="You are an assistant. Provide a class name for each image that can be used to name all images as assets."
    )


    content = []
    for img_url in image_urls:
        content.append(
            {
                "type": "image_url",
                "image_url": {"url": img_url},
                "detail": "low"
            }
        )


    with get_openai_callback() as cb:
        msg = chat.invoke(
            [
                system_message,
                HumanMessage(content=content)
            ]
        )

    print(cb)
    return msg.content

def passing_multiple_img(image_urls):
    """Pass multiple image URLs to the model and return the classifications."""

    class_name = image_class(image_urls)

    return  class_name

# Link must be in format of https://drive.google.com/uc?export=view&id= unique_id   (https://drive.google.com/file/d/1OBb305C4CXdY6zCpw8KY00jcDaHcLvwt/view?usp=drive_link) id =1OBb305C4CXdY6zCpw8KY00jcDaHcLvwt
image_urls = [
    "https://drive.google.com/uc?export=view&id=1V1W3-wpWElaNhUDcThLUbxr6NNauLf-z",
    "https://drive.google.com/uc?export=view&id=1OBb305C4CXdY6zCpw8KY00jcDaHcLvwt",


]

image_summaries = passing_multiple_img(image_urls)

print(image_summaries)
```
