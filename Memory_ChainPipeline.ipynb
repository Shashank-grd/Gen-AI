```python
import os
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
LANGCHAIN_API_KEY=userdata.get("LANGCHAIN_API_KEY")
```

```python
os.environ["LANGCHAIN_API_KEY"]=LANGCHAIN_API_KEY
os.environ["GOOGLE_API_KEY"]=GOOGLE_API_KEY
os.environ["LANGCHAIN_TRACING_V2"]="true"
```

```python
from langchain_community.document_loaders import TextLoader,DirectoryLoader
from langchain.memory import ConversationBufferMemory ,ConversationBufferWindowMemory
from langchain_google_genai import ChatGoogleGenerativeAI;
from langchain_core.prompts import ChatPromptTemplate;
from langchain.chains import ConversationChain
```

```python
loader = DirectoryLoader("data", glob="*.txt", loader_cls=TextLoader)
code = loader.load()
```

```python
print(code[0].page_content)
```

```python
memory=ConversationBufferWindowMemory(k=3)
```

```python
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","""You are a highly experienced and efficient software engineer specializing in code optimization and refactoring. Your goal is to take the input code provided by the user,just give code no explanation or etc and refactor it according to the following rules:
1. Ensure the code remains functionally equivalent after refactoring.
2. Simplify any complex or redundant logic.
3. Rename variables and functions to follow clear and consistent naming conventions.
4. Remove dead code, redundancies, or unnecessary elements.
5. Add concise, helpful comments where clarity is needed.
6. Optimize for performance without sacrificing readability.
7. Follow best practices and standards for the specific programming language  """),
        ("user","Code:{code}")
    ]
)
```

```python
def get_session_history():
    return memory.get_history()
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-1.5-flash")
chain= ConversationChain(llm=llm,memory=memory)
user_input = prompt.format(code=code)
response = chain.run(user_input)
print(response)
```
