```python
import os
import google.generativeai as genai
from PIL import Image
import cv2
import numpy as np

genai.configure(api_key="")

def get_gemini_response(input_prompt, image):
    model = genai.GenerativeModel('gemini-1.5-flash')

    token_prompt = model.count_tokens([input_prompt])
    print(token_prompt)
    token_image = model.count_tokens([image])
    print(token_image)

    response = model.generate_content([input_prompt, image])
    return response.text

def input_image_setup(file_path):
    try:
        image = Image.open(file_path)

        # Convert to grayscale if the image is not already
        if image.mode != 'L':
            image = image.convert('L')

        # Convert grayscale to RGB using OpenCV
        gray_image = np.array(image)
        colorized_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB)

        # Convert back to PIL Image for compatibility
        image = Image.fromarray(colorized_image)

        return image
    except FileNotFoundError:
        raise FileNotFoundError("File not found. Please check the file path.")

input_prompt = "getting these signals on my dashboardof my car, i hAVE PROVIDED THE IMAGE and car just broke down fter giving out black smoke, what should i do?"
uploaded_file_path = "/content/data/figure-12-30.jpg"

image_data = input_image_setup(uploaded_file_path)

response = get_gemini_response(input_prompt, image_data)

print(response)
```
