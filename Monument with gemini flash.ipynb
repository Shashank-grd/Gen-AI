```python
import google.generativeai as genai
from PIL import Image
```

```python
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-1.5-flash')
def get_gemini_response(input_prompt, image):

    token_prompt = model.count_tokens([input_prompt,image])
    print(token_prompt)
    #token_image = model.count_tokens([image])
    #print(token_image)

    response = model.generate_content([input_prompt, image])
    return response.text

def input_image_setup(file_path):
    try:
        image = Image.open(file_path)
        #image = image.resize((150, 150))
        return image
    except FileNotFoundError:
        raise FileNotFoundError("File not found. Please check the file path.")
```

```python
input_prompt = """You are an expert virtual tour guide. When shown an image of a monument, your task is to:
                     Recognize the monument from the image.
                     Provide the name of the monument.
                     Offer detailed information about the monument, including its historical significance, architectural style, and interesting facts.
                     Provide context on the location, cultural impact, and notable events related to the monument. """

uploaded_file_path = "/content/monu/hanuman tekri.jpg"

image_data = input_image_setup(uploaded_file_path)

response = get_gemini_response(input_prompt, image_data)

print(response)
```
