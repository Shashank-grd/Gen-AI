```python
LABEL_MAP_URL ="https://www.gstatic.com/aihub/tfhub/labelmaps/landmarks_classifier_asia_V1_label_map.csv"
```

```python
IMAGE_SHAPE = (321,321)
```

```python
df=pd.read_csv(LABEL_MAP_URL)
```

```python
df
```

```python
label_map=dict(zip(df.id,df.name))
label_map
```

```python
!pip install --upgrade tensorflow tensorflow-hub
```

```python
model_layer = hub.KerasLayer(TF_MODEL_URL, input_shape=IMAGE_SHAPE + (3,),  output_key="predictions:logits",trainable=False)
```

```python
import cv2
```

```python
def findImg(image):
  img=cv2.imread(image)
  resize=cv2.resize(img,(640,480))
  RGBimg = cv2.cvtColor(resize, cv2.COLOR_BGR2RGB)
  RGBimg =cv2.resize(img,(321,321))
  RGBimg = np.array(RGBimg)/255
  RGBimg = np.reshape(RGBimg,(1,321,321,3))
  predictions = model_layer(RGBimg)
  return label_map[np.argmax(predictions)]
```

```python
from langchain.prompts import PromptTemplate
from langchain import HuggingFaceHub,LLMChain
from transformers import AutoTokenizer
from langchain.llms import HuggingFacePipeline
import torch
import transformers
import os
```

```python
!huggingface-cli login
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model ="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer= AutoTokenizer.from_pretrained(model)
```

```python
pipeline= transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm=HuggingFacePipeline(pipeline=pipeline,model_kwargs={'temperature' : 0})
```

```python
prompt=PromptTemplate(
    input_variables=["monument"],
    template= """
      Give details about the {monument} in this format for a new person who dont know about the monument
      name:
      place:
      Famous for:
      details:
      ancient history:

       Only return the helpful answer below and nothing else.
""")
```

```python
chain =LLMChain(llm=llm,prompt =prompt)
```

```python
print(chain.run({"monument": result}))
```
